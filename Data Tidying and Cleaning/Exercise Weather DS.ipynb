```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_data = pd.read_csv("data/weather.csv")

# pd.set_option('display.max_rows', None) #
```

```python
"""
Here i melt my data from wide to long 
 - I will keep my columns that i dont want to change yet in id_vars=[..]
 - I will give var_name="day" and all these columns(d1, d2, d3, d4) will turn into one column under variable "day"
 - And these values of origin (d1, d2) columns will be under value_name="value" 
"""
weather_data = weather_data.melt(id_vars=["id", "year", "month", "element"], var_name="day")
weather_data
```

```python
# I remove all my NaN values from the 
# Specify in which column to look for missing value (by default) it is looking in all columns
# So in this case i am not required to provide subset but i just show that it works like that
weather_data = weather_data.dropna(subset=["value"])
```

```python
# I extract the number type of day by slicing and getting the all after 1 index of the string
# Then i convert this exctracted number into integer
weather_data.day = weather_data.day.str.slice(1).astype('int64') 
```

```python
# weather_data = weather_data.dropna() # I remove all NaN values looking in all the colums by default
weather_data = weather_data.dropna(subset="value") # I look for NaN values only in value column
```

```python
"""
From values in column `element` becomes `headers` of the new columns which i create with pivot_table
- values will become the values in the new columns
- index specify which columns i want to keep unchanged
"""
weather_data = weather_data.pivot_table(index=["id", "year", "month", "day"], columns="element", values="value")
```

```python
# Since pivot_table will return me DataFrame with multi indexes i need to fix that by reseting indexes
weather_data = weather_data.reset_index()
```

```python
# pd.to_datetime({'year': ['2000'], 'month': ['10'], 'day': ['10']}) This is how datetime works

"""
I want to create DateTime object with my information
"""
# I create simple data object that is going to be used with pandas method for datetime
object_with_dates = {
    'year': weather_data.year, # Collection of Series
    'month': weather_data.month,
    'day': weather_data.day
}

datetime_data_var = pd.to_datetime(object_with_dates)
```

```python
def convertor(x):
    # I am using a function to pass in apply to convert to datetime object
    dates = {
        'year': [x['year']],
        'month': [x['month']],
        'day': [x['day']]
    }
    return pd.to_datetime(dates)

# For every row i apply the function to create the datetime object
# And i create new column called `datetime`
weather_data['datetime'] = weather_data.apply(convertor, axis=1) 
```

```python
# After my datetime objects i dont need these columns anymore so i drop them
weather_data = weather_data.drop(columns=['year', 'month', 'day'])

# Then i take the subset so i can order columns as i want to
weather_data = weather_data[["id", "datetime", "tmax", "tmin"]]
```
