```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_data = pd.read_csv("https://raw.githubusercontent.com/synesthesiam/blog/master/posts/data/weather_year.csv")
```

```python
weather_data
```

```python
weather_data.columns
```

```python
# Rename the columns because some of them have start with Uppercase and have heading whitespace

weather_data.columns = ["date", "max_temp", "mean_temp", "min_temp", "max_dew",
 "mean_dew", "min_dew", "max_humidity", "mean_humidity",
 "min_humidity", "max_pressure", "mean_pressure",
 "min_pressure", "max_visibility", "mean_visibility",
 "min_visibility", "max_wind", "mean_wind", "max_gusts",
 "precipitation", "cloud_cover", "events", "wind_dir"]
```

```python
weather_data
```

```python
# I need my date to be in datetime object

weather_data.date = pd.to_datetime(weather_data.date)
```

```python
# Edit all fields that corresponds to this filter
# - loc locate by index first then by columns
weather_data.loc[weather_data.precipitation == "T", 'precipitation'] = 1e-6
```

```python
# This allows me to convert it to float now
weather_data.precipitation = weather_data.precipitation.astype(float)
```
